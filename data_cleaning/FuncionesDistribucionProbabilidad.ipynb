## Simulación de Monte Carlo

P(Caer en el circulo) = Casos favorables / Casos totales = A(circulo) / A(Cuadrado) = pi*r^2 / (2*r)^2 = pi / 4

#### Pasos para generar pi

1. generar dos números aleatorios entre 0 y 1.
2. calcular d = (x^2)+(y^2)
3. si d <= 1, entonces esta dentro del circulo. En cualquier otro caso, se encuentra fuera.
4. realizar los pasos 1, 2 y 3 multiples veces


```python
pi_avg = 0
n = 10000
reps = 200

for rep in range(reps):
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()

    casos_favorables = 0
    for i in range(n):
        if(x[i]*x[i] + y[i]*y[i] <= 1 ):
            casos_favorables += 1

    pi_avg += (4.0 * float(casos_favorables) / n)

print (pi_avg / reps)
```
